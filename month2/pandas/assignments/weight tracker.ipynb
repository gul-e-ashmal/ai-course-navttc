Handling Null values for Calories consumed

```python
df['Calories_Consumed']
df['Calories_Consumed'].isnull()
df['Calories_Consumed'].mean()
df['Calories_Consumed'].fillna(df['Calories_Consumed'].mean()) # did not make changes in table
df['Calories_Consumed']=df['Calories_Consumed'].fillna(df['Calories_Consumed'].mean())

```

```python
df['Calories_Consumed']
df
```

Handle missing value for workout-type

```python
# We will handle none value for workout-type beacause maybe people do not workout
df['Workout_Type']=df['Workout_Type'].fillna('None')
df['Workout_Type']
```

Insights for workout_type

```python
df['Workout_Type'].value_counts()
```

Check workout intensity should be zero for none value of workout type

```python
df[df['Workout_Type']=='None']
df[df['Workout_Type']=='None']['Workout_Intensity']
df[df['Workout_Type']=='None']['Workout_Intensity'].sum()
```

```python
# Annothe method --
df['Workout_Type']=='None'
df['Workout_Intensity'] == 0
df[(df['Workout_Type']=='None' )& (df['Workout_Intensity'] == 0)]['Workout_Intensity']
```

which workout type have max avg intensity

```python
df.groupby('Workout_Type')['Workout_Intensity'].mean()
max(df.groupby('Workout_Type')['Workout_Intensity'].mean())
df.groupby('Workout_Type')['Workout_Intensity'].mean().sort_values()
df.groupby('Workout_Type')['Workout_Intensity'].mean().sort_values(ascending=False)
df.groupby('Workout_Type')['Workout_Intensity'].mean().sort_values(ascending=False).head(1)
```

avg weight change of workout type

```python
df.groupby('Workout_Type')['Weight_Change'].mean()
max(df.groupby('Workout_Type')['Weight_Change'].mean())
df.groupby('Workout_Type')['Weight_Change'].mean().sort_values()
df.groupby('Workout_Type')['Weight_Change'].mean().sort_values(ascending=False)
```
